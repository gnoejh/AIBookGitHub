# Anthropic Model Context Provider (MCP)

This notebook contains implementation and examples for the Model Context Provider by Anthropic. The MCP is designed to manage context handling and interactions with Anthropic's AI models.

```python
import os
import anthropic

# Set up the client with your API key
client = anthropic.Anthropic(
    # Replace with your API key or set as an environment variable
    api_key=os.environ.get("ANTHROPIC_API_KEY"),
)

# Example function that provides context to the model
def get_response_with_context(user_query, context_documents):
    # Prepare the context by joining the documents
    context = "\n\n".join([doc for doc in context_documents])
    
    # Create the message with context
    message = client.messages.create(
        model="claude-3-opus-20240229",
        max_tokens=1024,
        system="You are a helpful assistant. Use the context provided to answer questions.",
        messages=[
            {
                "role": "user",
                "content": f"Context: {context}\n\nQuestion: {user_query}"
            }
        ]
    )
    
    return message.content

# Example usage
context_docs = [
    "The Model Context Provider (MCP) is a system that manages context for AI models.",
    "Anthropic's Claude models can process up to 200K tokens of context in a single query.",
    "Effective context management improves response quality and reduces hallucinations."
]

query = "How does the MCP help improve AI responses?"
response = get_response_with_context(query, context_docs)
print(response)
```
